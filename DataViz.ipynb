```python
import plotly.graph_objects as go

# Couleur personnalise plus fonce pour les barres
couleur_bars = 'rgba(0, 0, 128, 1)'  # Bleu fonc avec une opacit de 50%

# Crer la figure en utilisant les donnes correctes
fig = go.Figure(
    data=[
        go.Bar(
            x=df_note["sexe"],  # Utiliser les donnes de la colonne "sexe"
            y=df_note["note"],  # Utiliser les donnes de la colonne "note"
            marker=dict(color=couleur_bars)  # Appliquer la couleur aux barres
        )
    ],
    layout=go.Layout(
        title=go.layout.Title(text="Notes ngatives donnes par le sexe H ou F"),
        plot_bgcolor='white'  # Couleur de fond en blanc
    )
)

fig.show()
```

```python
import plotly.graph_objects as go

# Couleur personnalise pour les barres
couleur_bars = 'rgba(0, 0, 128, 1)'  # Bleu fonc

# Crer la figure en utilisant les donnes correctes
fig = go.Figure(
    data=[
        go.Bar(
            x=df_note[df_note["sexe"] == "H"]["nature"],  # Nature pour le sexe "H"
            y=df_note[df_note["sexe"] == "H"]["note"],    # Notes pour le sexe "H"
            marker=dict(color='blue'),  # Couleur pour le sexe "H"
            name='Homme'  # Lgende pour le sexe "H"
        ),
        go.Bar(
            x=df_note[df_note["sexe"] == "F"]["nature"],  # Nature pour le sexe "F"
            y=df_note[df_note["sexe"] == "F"]["note"],    # Notes pour le sexe "F"
            marker=dict(color='pink'),  # Couleur pour le sexe "F"
            name='Femme'  # Lgende pour le sexe "F"
        )
    ],
    layout=go.Layout(
        title=go.layout.Title(text="Notes ngatives donnes par la nature du produit"),
        plot_bgcolor='white',  # Couleur de fond en blanc
        width=1100,  # Largeur de la figure en pixels
        height=1000,  # Hauteur de la figure en pixels
        barmode='group'  # Mode de groupage des barres
    )
)

fig.show()
```
